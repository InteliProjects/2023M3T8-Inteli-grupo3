```python
# Define o ID do usuário-alvo
target_user_id = 1130135

# Lista das habilidades (skills) relevantes para recomendação
relevant_skills = ['improve_eng', 'improve_tech', 'improve_cur', 'improve_int', 'improve_dis', 'improve_neg', 'improve_sin', 'improve_res']

# Lista para armazenar as áreas de interesse do usuário-alvo
target_interest_areas = []

# Lista para armazenar os IDs das jornadas recomendadas
recommended_journey_ids = []

# Itera sobre as áreas para identificar as áreas de interesse do usuário-alvo
for area in areas:
    # Verifica se o usuário-alvo tem interesse na área atual
    is_target_interested = int(user_objectives[area][user_objectives['id'] == target_user_id])

    # Se o usuário tiver interesse (valor 1), adiciona a área à lista de áreas de interesse
    if is_target_interested == 1:
        target_interest_areas.append(area)

# Itera sobre as áreas de interesse do usuário-alvo
for target_area in target_interest_areas:
    # Classifica o DataFrame pela relevância dos cursos para a área alvo em ordem decrescente
    relevant_journey = df_relevance_courses[['journey_id', target_area]].sort_values(by=target_area, ascending=False)
    relevant_journey = relevant_journey[['journey_id']]

    # Itera sobre as habilidades relevantes
    for skill in relevant_skills:
        count = 0
        count_journey = 0
        # Lista para armazenar os IDs das jornadas já recomendadas nesta iteração
        recommended_journey_ids_iteration = []
        # Itera sobre os IDs das jornadas ordenadas por relevância para a área
        for journey_id in relevant_journey['journey_id']:
            # Itera sobre os IDs das jornadas que têm a habilidade atual (skill)
            for id_target_journey in df_cursos['journey_id'][df_cursos[skill] == 1]:
                for id_journey_user_target in df_user_journeys["journey_id"][df_user_journeys['user_id'] == target_user_id]:
                    if id_journey_user_target == id_target_journey:
                        count_journey += 1
                # Verifica se a jornada ainda não foi recomendada, se o limite de 10 recomendações não foi atingido e se a contagem atual é menor que 1
                if id_target_journey not in recommended_journey_ids and len(recommended_journey_ids) < 10 and count < 1 and count_journey < 1:
                    # Adiciona a jornada à lista de jornadas recomendadas e à lista de recomendações da iteração atual
                    recommended_journey_ids.append(id_target_journey)
                    recommended_journey_ids_iteration.append(id_target_journey)
                    count += 1

        # Verifica se já foram recomendados 8 cursos (todas as habilidades) nesta iteração
        if len(recommended_journey_ids_iteration) == 8:
            # Ordena as jornadas não recomendadas por relevância para a área
            not_recommended_journeys = df_relevance_courses[['journey_id', target_area]].sort_values(by=target_area, ascending=False)
            not_recommended_journeys = not_recommended_journeys[['journey_id']]

            # Itera sobre as jornadas não recomendadas
            for journey_id in not_recommended_journeys['journey_id']:
                # Verifica se a jornada não foi recomendada e se não está na lista de recomendações da iteração atual
                if journey_id not in recommended_journey_ids and journey_id not in recommended_journey_ids_iteration:
                    # Adiciona a jornada à lista de jornadas recomendadas e à lista de recomendações da iteração atual
                    recommended_journey_ids.append(journey_id)
                    recommended_journey_ids_iteration.append(journey_id)

                # Verifica se já foram recomendados 10 cursos
                if len(recommended_journey_ids) >= 10:
                    break

            # Encerra a iteração se já tiver recomendado 10 cursos
            if len(recommended_journey_ids) >= 10:
                break

# Imprime a lista de IDs das jornadas recomendadas
print(recommended_journey_ids)
```
